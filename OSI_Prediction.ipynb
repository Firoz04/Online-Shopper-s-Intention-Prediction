```python
from sklearn import preprocessing
scaler = preprocessing.StandardScaler().fit(X_res)
X1 = scaler.transform(X_res)

from sklearn.model_selection import train_test_split
X1,x2,y1,Y2 = train_test_split(X1,y_res,test_size=0.25)

from sklearn.metrics import f1_score
from sklearn.neighbors import KNeighborsClassifier
train = []
val = []
train_f1_score = []
val_f1_score = []
for i in range(1,21):
    neigh = KNeighborsClassifier(n_neighbors=i,algorithm='kd_tree',weights='distance')
    neigh.fit(X1, y1)
    train.append(neigh.score(X1,y1))
    val.append(neigh.score(x2,Y2))
    train_f1_score.append(f1_score(y1,neigh.predict(X1)))
    val_f1_score.append(f1_score(Y2,neigh.predict(x2)))
```

```python
plt.figure()
plt.title('Accuracy Metric')
plt.ylabel('Acc')
plt.xlabel('Number of K neigbours')
plt.plot(train)
plt.plot(val)
plt.figure()
plt.title('F1 Score Metric')
plt.ylabel('f1 score')
plt.xlabel('Number of K neigbours')
plt.plot(train_f1_score)
plt.plot(val_f1_score)
```

```python
y = df['Revenue'].copy()
x = df.drop('Revenue', axis = 1)
```

```python
#Support Vector Machine
x_train_SVM = x_train
x_test_SVM = x_test
y_train_SVM = y_train
y_test_SVM = y_test
```

```python
pd.DataFrame(x_train_SVM)
```

```python
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
from sklearn.metrics import roc_auc_score, roc_curve, classification_report, precision_recall_curve, auc
from sklearn.svm import SVC
svc = SVC(kernel ='rbf')
svc.fit(x_train_SVM, y_train_SVM)
y_pred_SVM=svc.predict(x_test_SVM)

print("Accuracy: \n", accuracy_score(y_test_SVM, y_pred_SVM))
print("Classification Report: \n", classification_report(y_test_SVM, y_pred_SVM))
#curve
fpr_SVM, tpr_SVM, threshold = roc_curve(y_test_SVM, y_pred_SVM)
svm_roc = (fpr_SVM, tpr_SVM)
print('Support Vector Machine ROC = ',(svm_roc))
#area under the curve
svm_auc = roc_auc_score(y_test_SVM, y_pred_SVM)
print("Support Vector Machine AUC = %.3f" % (svm_auc))
```

```python
y_head_svc = svc.predict(x_test_SVM)
```

```python
from sklearn.metrics import confusion_matrix

Cm_svc = confusion_matrix(y_test_SVM,y_head_svc)
```
